```python
import pandas as pd
```

```python
import pyodbc
server = '' 
database = '' 
username = '' 
password = ''
cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER='+server+';DATABASE='+database+';UID='+username+';PWD='+ password)
cursor = cnxn.cursor()
```

```python
# DF ALL DAP & KCL FROM Pengadaan_ME2N
df_all_leadtime = pd.read_sql_query("SELECT * FROM (SELECT Pengadaan_ME2N.BANFN,Pengadaan_ME2N.BNFPO,Pengadaan_ME2N.EBELN,Pengadaan_ME2N.EBELP, Pengadaan_EKBE.BUDAT,Pengadaan_ME5A_R1.BADAT, Pengadaan_ME2N.EMATN, Pengadaan_ME2N.TXZ01, Pengadaan_ME2N.WERKS, BDA_MARC.BSTMI AS LOTSIZE, BDA_MARC.EISBE AS SAFETYSTOCK FROM Pengadaan_ME2N LEFT JOIN Pengadaan_ME5A_R1 ON Pengadaan_ME2N.BANFN = Pengadaan_ME5A_R1.BANFN AND Pengadaan_ME2N.BNFPO = Pengadaan_ME5A_R1.BNFPO LEFT JOIN Pengadaan_EKBE ON Pengadaan_EKBE.EBELN = Pengadaan_ME2N.EBELN AND Pengadaan_EKBE.EBELP = Pengadaan_ME2N.EBELP LEFT JOIN BDA_MARC ON BDA_MARC.MATNR = Pengadaan_ME2N.EMATN AND BDA_MARC.WERKS = Pengadaan_ME2N.WERKS WHERE (Pengadaan_EKBE.BWART = '101' OR Pengadaan_EKBE.BWART = '103') AND Pengadaan_ME2N.EMATN LIKE '000000000006%' AND Pengadaan_ME2N.BANFN IS NOT NULL AND BDA_MARC.MAABC = 'A') AS TBL1 WHERE BADAT IS NOT NULL ORDER BY BANFN, BNFPO, EBELN",cnxn)
```

```python
df_c = df_all_leadtime
df_c
```

```python
# buat cek duplicate
df_c[df_c.duplicated(subset=['BANFN','BNFPO','EBELN','EBELP'])]
```

```python
df_c.drop_duplicates(subset =['BANFN','BNFPO','EBELN','EBELP','BUDAT','BADAT'],inplace=True)
```

```python
df_c['BUDAT'] = df_c['BUDAT'].str[:4] + '-' + df_c['BUDAT'].str[4:6] + '-' + df_c['BUDAT'].str[6:8]
df_c['BADAT'] = df_c['BADAT'].str[:4] + '-' + df_c['BADAT'].str[4:6] + '-' + df_c['BADAT'].str[6:8]
```

```python
df_c
```

```python
df_c['BUDAT'] = pd.to_datetime(df_c['BUDAT'])
```

```python
df_c.drop(df_c[df_c['BADAT'] ==  '4752-12-08'].index, inplace=True)
```

```python
df_c['BADAT'] = pd.to_datetime(df_c['BADAT'])
```

# Cek duplicate tanggal dan ambil tanggal max

```python
df_c[df_c['BANFN'] == '2200000056']
```

```python
# Untuk BUDAT kata mas ghifari diambil nilai min
# kenapa ga max.
c = df_c.groupby(['BANFN','BNFPO','EBELN','EBELP']).agg({'BUDAT' : 'max'}).reset_index()
```

```python
final = pd.merge(df_c,c,on=['BANFN','BNFPO','EBELN','EBELP','BUDAT'])
```
